```python
# Project Coachbot
# Mauricio Ruiz Díaz
# Matriculation Number: 38149112

# Import necessary libraries
# This imports the math module, which provides mathematical functions where you can add them later.
import math

# Create a dictionary to store user data
# This dictionary stores specific information such as name, weight, height, age, activity level, points earned, and current level. 
# It acts as a central place to manage user data. 
user_data = {
    "name": "",
    "weight_kg": 0,
    "height_cm": 0,
    "age": 0,
    "activity_level": "",  # Sedentary, Lightly Active, Moderately Active, Very Active
    "points": 0,
    "level": 1
}

# Create a dictionary to store training plans, each plan is a list of activities.
training_plans = {
    "beginner": ["10-minute warm-up", "20-minute walk", "10-minute stretch"],
    "intermediate": ["15-minute warm-up", "30-minute jog", "15-minute strength training"],
    "advanced": ["20-minute warm-up", "45-minute run", "20-minute intense strength training"]
}

# Function to calculate BMI according to "Australian Heart Fundation" 
#It converts height from centimeters to meters and rounds the result to 2 decimal places.

def calculate_bmi(weight_kg, height_cm):
    height_m = height_cm / 100
    bmi = weight_kg / (height_m ** 2)
    return round(bmi, 2)

# Function to calculate daily calorie needs
# his function estimates Mifflin-St Jeor equation for calculate the calories and adjusts it based on the user's activity level.
def calculate_calories(weight_kg, height_cm, age, activity_level):
    # Basal Metabolic Rate (BMR) using Mifflin-St Jeor equation
    bmr = 10 * weight_kg + 6.25 * height_cm - 5 * age + 5  # For males
    # Adjust BMR based on activity level
    activity_factors = {
        "sedentary": 1.2,
        "lightly active": 1.375,
        "moderately active": 1.55,
        "very active": 1.725
    }
    calories = bmr * activity_factors.get(activity_level.lower(), 1.2)
    return round(calories)

# Function to display training plan
# This function displays the training plan for a given fitness level (beginner, intermediate, advanced). If the level is invalid, it notifies the user.
def display_training_plan(level):
    plan = training_plans.get(level.lower(), [])
    if plan:
        print(f"Your {level} training plan:")
        for i, activity in enumerate(plan, start=1):
            print(f"{i}. {activity}")
    else:
        print("Invalid training level. Choose from beginner, intermediate, or advanced.")

# Function to add points and check for level up
# This function adds points to the user's total and checks if they have earned enough points to level up. The user levels up every 100 points.
def add_points(points):
    user_data["points"] += points
    print(f"You earned {points} points! Total points: {user_data['points']}")
    
    # Check for level up (every 100 points)
    if user_data["points"] >= user_data["level"] * 100:
        user_data["level"] += 1
        print(f"Congratulations! You leveled up to Level {user_data['level']}!")

# Main program loop
# This is the main function that runs the program
def main():

    # Welcoms the program
    print("Welcome to Coachbot! Let's get started.")
    
    # Collect user information
    user_data["name"] = input("Enter your name: ")
    user_data["weight_kg"] = float(input("Enter your weight in kg: "))
    user_data["height_cm"] = float(input("Enter your height in cm: "))
    user_data["age"] = int(input("Enter your age: "))
    user_data["activity_level"] = input("Enter your activity level (Sedentary/Lightly Active/Moderately Active/Very Active): ")

    # Provides a menu for the user to choose from:
    while True:
        print("\nWhat would you like to do?")
        print("1. Check BMI")
        print("2. Check Daily Calorie Needs")
        print("3. View Training Plan")
        print("4. Earn Points")
        print("5. Exit")
        
        choice = input("Choose an option (1/2/3/4/5): ")
        
        if choice == "1":
            bmi = calculate_bmi(user_data["weight_kg"], user_data["height_cm"])
            print(f"Your BMI is {bmi}.")
            if bmi < 18.5:
                print("You are underweight.")
            elif 18.5 <= bmi < 24.9:
                print("You are at a healthy weight.")
            elif 25 <= bmi < 29.9:
                print("You are overweight.")
            else:
                print("You are obese.")
        
        elif choice == "2":
            calories = calculate_calories(user_data["weight_kg"], user_data["height_cm"], user_data["age"], user_data["activity_level"])
            print(f"Your estimated daily calorie needs are {calories} calories.")
        
        elif choice == "3":
            level = input("Enter your training level (Beginner/Intermediate/Advanced): ")
            display_training_plan(level)
        
        elif choice == "4":
            points = int(input("Enter the number of points you earned (e.g., for completing a workout): "))
            add_points(points)
        
        elif choice == "5":
            print("Exiting Coachbot. Goodbye!")
            break
        
        else:
            print("Invalid choice. Please try again.")
            
        # Handles user input and calls the appropriate functions.

# Run the program
if __name__ == "__main__":
    main()
```
